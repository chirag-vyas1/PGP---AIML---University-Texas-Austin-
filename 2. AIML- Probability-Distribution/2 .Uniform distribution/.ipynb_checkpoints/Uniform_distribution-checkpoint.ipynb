# **`1. Theory- Uniform Distribution`**

The **Uniform distribution** is a type of probability distribution where **all outcomes are equally likely** within a given range.

There are **two main types** of Uniform distributions:
1. **Discrete Uniform Distribution**
2. **Continuous Uniform Distribution**

---

## 1. Discrete Uniform Distribution

- Each outcome has **equal probability**.
- Used when there are a **finite number of outcomes**.
- Example: Rolling a fair die (1 to 6) — each number has probability \( \frac{1}{6} \).

**Formula:**
\[
P(X = x) = \frac{1}{n}
\]
where \( n \) is the total number of equally likely outcomes.

---

## 2. Continuous Uniform Distribution

- Used when outcomes can take **any value in a continuous interval** \([a, b]\).
- The probability density is **constant** across the interval.

**Probability Density Function (PDF):**
\[
f(x) =
\begin{cases}
\frac{1}{b - a}, & a \le x \le b \\
0, & \text{otherwise}
\end{cases}
\]

**Mean:**
\[
E[X] = \frac{a + b}{2}
\]

**Variance:**
\[
Var[X] = \frac{(b - a)^2}{12}
\]

---

### Key Features:
| Feature | Description |
|----------|--------------|
| **Shape** | Flat / Rectangular |
| **Parameters** | \( a \) (minimum), \( b \) (maximum) |
| **Range of values** | Between \( a \) and \( b \) |
| **Probability density** | Constant across the interval |
| **Used for** | Modeling random events where each outcome is equally likely |

---

###  Examples
- Rolling a fair die (Discrete case)
- Randomly choosing a number between 0 and 1 (Continuous case)
- Selecting a random time between 2 PM and 4 PM

---

###  Intuition:
If you **don’t prefer one outcome over another** — like picking a random point on a line segment —  
then you’re dealing with a **Uniform distribution**.

Another way to recognize a uniform distribution in your data is to look at a density plot. We will use displot of seaborn library to visualize the distribution of time needed for bug fixing.

**Insight**: As you can see from the above plot that all the values between 1 and 5 are having almost equal probability, we are going to use continuous uniform distribution. We need to decide the endpoints. Here, endpoints are 1 and 5.

X ~ U(1, 5)

````python
#import the required function
from scipy.stats import uniform

# use the uniform .pmf() function to generate the probabilty distribution
x = np.linspace(1,5,50)
prob = uniform.pdf(x,loc =1, scale =4)
````

**Find the probability that a randomly selected software debugging requires a maximum time of 3 hours** 

**CDF:** of a random variable (X) is the probability that X  will take the value less than or equal to x. It can be represented mathematically as below.

>$F_X(x) = P(X\leq x)$


In our case, random variable (X) is the number of hours.

$ P(X\leq 3)$

````python
# plot the probability distribution
# We are plotting the distributions here to better visualize the calculations.
# Of course you do not 'need' to create the following visualization to answer the question above.
# You can directly use the cdf function for probability calculations.

probs = uniform.pdf(x= x,loc=1,scale=4)
plt.plot(x, probs)
x1 = np.linspace(1, 3, 25)
plt.fill_between(x, probs,color="green")
plt.fill_between(x1, uniform.pdf(x=x1, loc=1, scale=4), color="r")
plt.xlabel("Time required for bug fixing")
plt.ylabel("Probability")
plt.title("Continuous Uniform Distribution: X ~ U(1,5)")
plt.show()
````

In the above graph, the red region represents P(X<=3). Let's calculate the probability that that a randomly selected software debugging requires a maximum time of 3 hours. We will use uniform.cdf() for this.

````python
uniform.cdf(x=3, loc=1, scale=4)
````


**Find the probability that a randomly selected software bug fixing requires more than two hours.**

$ P(X>2)$

````python
# plot the probability distribution
# We are plotting the distributions here to better visualize the calculations.
# Of course you do not 'need' to create the following visualization to answer the question above.
# You can directly use the cdf function for probability calculations.
x1 = np.linspace(2, 5, 20)
plt.plot(x, probs)
plt.fill_between(x, probs)
plt.fill_between(x1, uniform.pdf(x=x1, loc=1, scale=4), color="r")
plt.xlabel("Time required for bug fixing")
plt.ylabel("Probability")
plt.title("Continuous Uniform Distribution: X ~ U(1,5)")
plt.show()
````

In the above graph, the reg region represent P(X>2). Let's calculate the probability that that that a randomly selected software debugging requires more than two hours. We will use uniform.cdf() for this.

````python
1 - uniform.cdf(x=2, loc =1 , scale =4)
````

**Let's calculate the 50th percentile of software debugging time.**

ppf(): It is used to calculate the percentile point given probability. It works opposite of cdf()

````python
uniform.ppf(q=0.5, loc=1, scale=4)
````

### Conclusion:  

There is a 50% chance that a randomly selected software debugging requires less than three hours.

There is a 75% chance that a randomly selected software debugging requires more than two hours.

The 50th percentile of the software debugging time is 3 hours.

A call center records the time (in minutes) a customer spends on a call, ranging uniformly between 2 and 10 minutes.  

Let:

X is a continuous random variable


**Questions:**
1. Find the probability that a randomly selected call lasts **less than 5 minutes**.  
2. Find the probability that a randomly selected call lasts **more than 7 minutes**.  
3. Find the **median (50th percentile)** of call duration.  






````python
# --------------------------- Creating fictitious data set ---------------------------
# Example 1: Customer Service Call Duration
# Uniform between 2 and 10 minutes
n_calls = 1000
call_duration = np.random.uniform(low=2, high=10, size=n_calls)
df_calls = pd.DataFrame({"CallDuration_min": call_duration})
df_calls.to_csv("customer_calls.csv", index=False)

# ---------------------------------------------------------------------------------
````

````python
df_calls.head()
````

````python
x=np.linspace(2,10)
````

````python
plt.hist(df_calls["CallDuration_min"],density=True)
plt.axhline(0.12, color="red")
plt.xlabel("call duration")
plt.ylabel("Probability")
plt.title("Data Distribution")
plt.show()
````

````python
# Density plot of time taken to fix the bug
sns.displot(df_calls["CallDuration_min"], kde=True)
plt.show()
````

````python
# import the required function
from scipy.stats import uniform

# use the uniform.pmf() function to generate the probability distribution
x = np.linspace(2, 10, 50)
probs = uniform.pdf(x, loc=2, scale=8)
````

````python
# plot the probability distribution
# We are plotting the distributions here to better visualize the calculations.
# Of course you do not 'need' to create the following visualization to answer the question above.
# You can directly use the cdf function for probability calculations.
x1 = np.linspace(2,5, 25)
plt.plot(x, probs)
plt.fill_between(x, probs)
plt.fill_between(x1, uniform.pdf(x=x1, loc=2, scale=8), color="r")
plt.xlabel("call duration")
plt.ylabel("Probability")
plt.title("Continuous Uniform Distribution: X ~ U(2,10)")
plt.show()
````

**1. Find the probability that a randomly selected call lasts less than 5 minutes.**

````python
uniform.cdf(x=5, loc =2, scale=8)
````

**2.Find the probability that a randomly selected call lasts more than 7 minutes.**

````python
1- uniform.cdf(x=7, loc =2, scale=8)
````

**3.Find the median (50th percentile) of call duration.**

````python
uniform.ppf(q=0.5,loc=2,scale=8)
````

## **`Conclusion`**

From the analysis of the continuous uniform distribution for software debugging time, customer call durations, delivery times, and problem-solving times, we can conclude the following:

1. **Uniform distributions assume all outcomes in a given range are equally likely.**  
   - For example, any call duration between 2 and 10 minutes is equally probable.

2. **Probability calculations are straightforward using the CDF and PPF functions.**  
   - `cdf(x)` gives the probability that the random variable is less than or equal to `x`.  
   - `ppf(q)` gives the value corresponding to a given cumulative probability `q` (e.g., median, percentile).

3. **Percentiles divide the range of possible outcomes evenly.**  
   - The median (50th percentile) splits the probability range into two equal halves.

4. **Practical insight:**  
   - For uniform distributions, simple arithmetic can also verify probabilities: \( P(a \le X \le b) = \frac{b - a}{\text{range}} \).  
   - This makes uniform distributions ideal for modeling **equally likely outcomes** in real-life scenarios such as time durations, delivery periods, or random sampling events.

5. **Visualizations (plots) help understand probabilities and intervals under the curve.**  
   - Highlighting specific intervals (like red areas for X between 1–3 hours) makes it easier to interpret real-world probabilities.

Overall, the uniform distribution is a simple yet powerful tool for modeling and analyzing scenarios where **every outcome in a range is equally likely**, providing both visual and quantitative understanding of the data.


## <a id=link4><a/>**`Example 3: Delivery Time of Online Orders`**
An e-commerce company records delivery times for orders in a city. Delivery times are **uniformly distributed** between 1 and 6 days.  

Let: 
X = Delivery time (days)
X is a continuous random variable
**Questions:**
1. Find the probability that an order is delivered in **less than 3 days**.  
2. Find the probability that an order takes **more than 4 days**.  
3. Find the **25th percentile** of delivery time.  

````python
# -------------------------------
# Example 2: Delivery Time of Online Orders
# Uniform between 1 and 6 days
n_orders = 1000
delivery_time = np.random.uniform(low=1, high=6, size=n_orders)
df_delivery = pd.DataFrame({"DeliveryTime_days": delivery_time})
df_delivery.to_csv("delivery_times.csv", index=False)

# -------------------------------
````

````python
df_delivery.head()
````

````python
delivery_days= np.linspace(1,6,50)
plt.hist(df_delivery["DeliveryTime_days"],density=True)
plt.xlabel("num of days")
plt.axhline(0.18,color="r")
plt.ylabel("probability")
plt.title("data distribution")
plt.show()
````

**1. Find the probability that an order is delivered in less than 3 days**.  

````python
x=np.linspace(1,6,50)
uniform_dist=uniform.pdf(x=x,loc=1,scale=5)
plt.fill_between(x,uniform_dist,color="blue")
x1=np.linspace(1,3,25)
uniform_dist_cdf=uniform.pdf(x1,loc=1,scale=5)
plt.fill_between(x1,uniform_dist_cdf,color="green")
plt.show()
````

````python
uniform_dist_cdf = uniform.cdf(x=3,loc=1,scale=5)
print(uniform_dist_cdf)
````

**2.Find the probability that an order takes more than 4 days.**

````python
1-uniform.cdf(x=4,loc=1,scale=5)
````

**3. Find the **25th percentile** of delivery time.**  

````python
uniform.ppf(q=0.25,loc=1,scale=5)
````

## **`Conclusion`**

- Delivery times are uniformly distributed between 1 and 6 days.
- 25% of orders are delivered within 2.25 days.
- The probability that an order is delivered in less than 3 days is 40%.
- The scale in the uniform distribution represents the total range of delivery times (6 − 1 = 5), not the specific day we are calculating the probability for.

## <a id=link5><a/>**`Example 4: Time to Solve a Math Problem`**
In a math competition, students take a **uniformly distributed** amount of time between 5 and 15 minutes to solve a problem.  

Let:  
X = Time to solve the problem (minutes)
X is a continuous random variable

**Questions:**
1. Find the probability that a student solves the problem in **less than 8 minutes**.  
2. Find the probability that a student takes **more than 12 minutes**.  
3. Find the **75th percentile** of time to solve the problem.  


````python
# -------------------------------
# Example 3: Time to Solve a Math Problem
# Uniform between 5 and 15 minutes
n_students = 1000
solve_time = np.random.uniform(low=5, high=15, size=n_students)
df_solve = pd.DataFrame({"SolveTime_min": solve_time})
df_solve.to_csv("math_solve_times.csv", index=False)

print("CSV files created successfully!")
````

````python
df_solve.head()
````

````python
plt.hist(df_solve['SolveTime_min'],density=True)
plt.title("Data distribution")
plt.xlabel("Time require to solve maths sum")
plt.axhline(0.09,color="r")
plt.ylabel("probability")
plt.show()
````

**1.Find the probability that a student solves the problem in less than 8 minutes.**

````python
x=np.linspace(5,15,50)
uniform_dist=uniform.pdf(x,loc=5,scale=10)
plt.fill_between(x,uniform_dist,color="b")
x1=np.linspace(5,8,25)
plt.fill_between(x1,uniform.pdf(x1,loc=5,scale=10),color="r")
plt.title("Data distribution")
plt.xlabel("Time require to solve maths sum")
plt.ylabel("probability")
plt.show()

````

````python
uniform.cdf(x=8,loc=5,scale=10)
````

**2. Find the probability that a student takes more than 12 minutes.**


````python
x=np.linspace(5,15,50)
#note just for visualization we are using pdf for plotting while calculationg we have to use cdf
uniform_dist=uniform.pdf(x,loc=5,scale=10)
plt.fill_between(x,uniform_dist,color="b")
x1=np.linspace(12,15,25)
plt.fill_between(x1,uniform.pdf(x1,loc=5,scale=10),color="r")
plt.title("Data distribution")
plt.xlabel("Time require to solve maths sum")
plt.ylabel("probability")
plt.show()
````

````python
1 - uniform.cdf(x=12,loc=5,scale=10)
````

**3. Find the 75th percentile of time to solve the problem.**

````python
uniform.ppf(q=.75,loc=5,scale=10)
````

## **`Conclusion`**

From the uniform distribution \( X \sim U(5,15) \), we observed that every value within the range 5 to 15 hours is equally likely for problem-solving time.  
- The probability that the problem is solved in less than 8 hours is **0.30**, meaning there’s a 30% chance of completing it early.  
- The probability that it takes more than 12 hours is also **0.30**, showing a symmetric distribution around the midpoint.  
- The **75th percentile (12.5 hours)** indicates that 75% of the problems are solved within 12.5 hours.

Thus, the uniform distribution provides a simple way to understand equally likely outcomes within a given range and helps in estimating completion probabilities and percentiles effectively.
